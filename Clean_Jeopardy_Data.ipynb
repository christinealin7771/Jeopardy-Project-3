```python
import pandas as pd

indexes = list()

raw_questions = pd.read_csv('JEOPARDY_CSV.csv',)
raw_questions.columns = raw_questions.columns.str.strip().copy()
raw_questions = raw_questions[raw_questions.Round != 'Final Jeopardy!'].copy()
#display(raw_questions)

trimmed_questions = raw_questions.nlargest(n=100000, columns=['Show Number']).copy()
trimmed_questions.reset_index(inplace = True)
#display(trimmed_questions)

clean_questions = trimmed_questions[['Air Date','Category','Value','Question','Answer']].copy()
#display(clean_questions)

questions_2012 = clean_questions[clean_questions['Air Date'].str.contains('2012')].copy()
questions_2012.reset_index(inplace = True)
display(questions_2012)

questions_2011 = clean_questions[clean_questions['Air Date'].str.contains('2011')].copy()
questions_2011.reset_index(inplace = True)
display(questions_2011)

questions_2010 = clean_questions[clean_questions['Air Date'].str.contains('2010')].copy()
questions_2010.reset_index(inplace = True)
display(questions_2010)

questions_2009 = clean_questions[clean_questions['Air Date'].str.contains('2009')].copy()
questions_2009.reset_index(inplace = True)
display(questions_2009)

questions_2008 = clean_questions[clean_questions['Air Date'].str.contains('2008')].copy()
questions_2008.reset_index(inplace = True)
display(questions_2008)

questions_2007 = clean_questions[clean_questions['Air Date'].str.contains('2007')].copy()
questions_2007.reset_index(inplace = True)
display(questions_2007)

questions_2006 = clean_questions[clean_questions['Air Date'].str.contains('2006')].copy()
questions_2006.reset_index(inplace = True)
display(questions_2006)

questions_2005 = clean_questions[clean_questions['Air Date'].str.contains('2005')].copy()
questions_2005.reset_index(inplace = True)
display(questions_2005)

questions_2004 = clean_questions[clean_questions['Air Date'].str.contains('2004')].copy()
questions_2004.reset_index(inplace = True)
display(questions_2004)
```

```python
clean_questions.to_csv('Jeopardy_cleaned_csv.csv')
clean_questions.to_csv('Jeopardy_cleaned_tsv_updated.tsv', sep='\t')
clean_questions.to_json('Jeopardy_cleaned_json.json')

questions_2012.to_csv('2012_questions.tsv', sep='\t')
questions_2011.to_csv('2011_questions.tsv', sep='\t')
questions_2010.to_csv('2010_questions.tsv', sep='\t')
questions_2009.to_csv('2009_questions.tsv', sep='\t')
questions_2008.to_csv('2008_questions.tsv', sep='\t')
questions_2007.to_csv('2007_questions.tsv', sep='\t')
questions_2006.to_csv('2006_questions.tsv', sep='\t')
questions_2005.to_csv('2005_questions.tsv', sep='\t')
questions_2004.to_csv('2004_questions.tsv', sep='\t')
```
